# We'll be generating an average_rating(score) through NLTK by calculating the sentiment score for individual reviews and 
# finding the average of all these reviews which will be the average rating corresponding to the movie.

# NLTK(Natural Language Toolkit): is a Python library for working with human language data. It provides a range of tools for tasks such as tokenization, stemming, tagging, parsing, and sentiment analysis .

# Vader(Valence Aware Dictionary & Sentiment Reasoner): is a part of the NLTK library that is specifically designed for sentiment analysis of social media content, news articles, and other types of text data that contain informal language, slang, and other nuances. VADER uses a lexicon (a list of words and their associated sentiment scores) and a set of grammar rules to determine the sentiment polarity (positive, negative, or neutral) and intensity (strength) of each sentence in a given text.

# VADER uses the following steps to calculate a sentiment score for a sentence: 

# 1) Tokenize the sentence into individual words.
# 2) Check each word in the lexicon and assign it a sentiment score (positive, negative, or neutral) based on its context and the words around it.
# 3) Apply a set of grammar rules to adjust the sentiment score based on the structure of the sentence (e.g., negations, conjunctions, intensifiers).
# 4)Calculate the overall sentiment score for the sentence by aggregating the scores of all the individual words and applying a normalization function.










## Sentiment Score Generator Code: 

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Create a new column to hold the average sentiment score for each movie
movie_reviews['average_rating'] = 0.0

# Create an instance of the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Loop over the rows of the dataframe.'i' is the index of the row, and 'row' is the data in that row.
for i, row in movie_reviews.iterrows():
    
    # Initialize a list to hold the individual review scores for this movie
    review_scores = []
    
    # This line starts a nested loop that iterates over the first 100 columns in the current row. 
    # Each column represents a movie review.
    for j in range(100):
        
        # Get the text of the review
        review_text = str(row[j+1])
        
        # Calculate the sentiment score for the review using VADER.
        # The 'compound' score is used, which is a normalized score between -1 (most negative) and +1 (most positive).
        review_score = sid.polarity_scores(review_text)['compound']
        
        # Append the score to the list of review scores
        review_scores.append(review_score)
        
    # Calculate the average of the review scores for this movie
    average_score = sum(review_scores) / len(review_scores)
    
    # Assign the average score to the new column for this row.The average sentiment score multiplied by 2.5 
    # and shifted by 1 to map the score to the [0, 5] range.
    movie_reviews.at[i, 'average_rating'] = (average_score + 1) * 2.5  # map score to [0,5]
```
