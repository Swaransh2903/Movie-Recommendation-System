# In this step , we extract the unique movies because we have to store them in separate csv.Why? Currently our dataset 'final_dataset(2).csv' contains columns like rating,userId,original_title and there are large number of instances of original_title of movie corresponding to every user who has seen the movie and in order to generate reviews corresponding to every movie for further analysis, we'll have to generate the unique movie titles.

```python
unique_movies_df=pd.DataFrame(user_movie_rating_df["original_title"].unique(),columns=["unique_original_title"])
```

```python
unique_movies_df
```

```python
unique_movieId_df=pd.DataFrame(user_movie_rating_df["movieId"].unique(),columns=["unique_movieId"])
```
