```python
import numpy as np 
import pandas as pd
```

```python
complete_df = pd.read_csv(r'complete-dataset(4).csv')
complete_df = complete_df.drop(columns='Unnamed: 0',axis=1)
complete_df
```

```python
ratings_matrix = complete_df.pivot_table(index='userId',columns='movieId',values='rating').fillna(0)
ratings_matrix
```

```python
import pandas as pd
nltk_ratings_matrix = pd.pivot_table(complete_df, values='nltk_rating', index='userId', columns='movieId').fillna(0)
nltk_ratings_matrix
```

# Recommendations based on Cosine Similarity b/w users:

## (When user leniency score is not taken into account):

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim_matrix = cosine_similarity(ratings_matrix)
cosine_sim_matrix_df = pd.DataFrame(index=complete_df['userId'].unique(),columns=complete_df['userId'].unique())
cosine_sim_matrix_df.values[:,:] = cosine_sim_matrix
cosine_sim_matrix_df
```

## -> Use K Nearest Neighbour for finding top 5 similar users based on cosine similarity:

```python
from sklearn.neighbors import NearestNeighbors

# fit a k-NN model on cosine similarity matrix
knn_model = NearestNeighbors(n_neighbors=6, metric='cosine', algorithm='brute')
knn_model.fit(cosine_sim_matrix_df)

def recommend_movies(user_id):
    
    # find the k-nearest neighbors of the user:
    _, indices = knn_model.kneighbors(cosine_sim_matrix_df.loc[[user_id]])
    similar_users = cosine_sim_matrix_df.iloc[indices[0], 1:]
    
    # get the movies not watched by the user:
    not_watched = ratings_matrix.loc[user_id][ratings_matrix.loc[user_id] == 0].index
    
    # create a dictionary to store the score of each not watched movie
    movie_scores = {}
    
    # iterate over each similar user
    for user in similar_users.index:
        
        # get the ratings of the user for not watched movies
        user_ratings = ratings_matrix.loc[user][not_watched]
        
        # calculate the cosine similarity score between the user and input user
        similarity_score = cosine_sim_matrix_df[user_id][user]
        
        # multiply the ratings with the similarity score and add to the movie_scores dictionary
        for movie, rating in user_ratings.items():
            if movie in movie_scores:
                movie_scores[movie] += rating * similarity_score
            else:
                movie_scores[movie] = rating * similarity_score
    
    # sort the movie_scores dictionary by value in descending order
    sorted_scores = sorted(movie_scores.items(), key=lambda x: x[1], reverse=True)
    
    # get the top 5 movies from the sorted_scores
    top_5_movies = [x[0] for x in sorted_scores[:5]]
    
    return top_5_movies
    

userId = 1
top_5_recommended_movies = recommend_movies(userId)
movie_titles = complete_df.loc[complete_df['movieId'].isin(top_5_recommended_movies), 'original_title'].tolist()
print(f"Recommended movies for {userId} are: ")
print()
unique_movie_titles = set(movie_titles)
for titles in unique_movie_titles:
    print(titles)

```

## (When user leniency score is taken into account):

```python
# importing the csv that contains user leniency matrix created by taking into account the leniency score of each user
user_leniency_matrix = pd.read_csv(r'user_leniency_matrix.csv')
user_leniency_matrix = user_leniency_matrix.drop(columns='Unnamed: 0',axis=1)
user_leniency_matrix.index = user_leniency_matrix.columns
user_leniency_matrix
```

## To make the hybrid recommendation based on both user ratings and leniency score, first we create a new matrix ,whose values will be obtained by giving 55% weightage to cosine similarity (b/w that pair of users) and 45% weightage to leniency score of the user:

```python
combined_score_matrix_cosine_and_leniency = 0.55 * cosine_sim_matrix + 0.45 * user_leniency_matrix
combined_score_matrix_cosine_and_leniency_df = pd.DataFrame(index=complete_df['userId'].unique(),columns=complete_df['userId'].unique())
combined_score_matrix_cosine_and_leniency_df.values[:,:] = combined_score_matrix_cosine_and_leniency
combined_score_matrix_cosine_and_leniency_df
```

## -> Use K Nearest Neighbour for finding top 5 similar users based on cosine similarity + leniency score:

```python
from sklearn.neighbors import NearestNeighbors

knn_model = NearestNeighbors(n_neighbors=6, metric='cosine', algorithm='brute')
knn_model.fit(combined_score_matrix_cosine_and_leniency_df)

# function to get top 5 movies based on similar users
def recommend_movies(user_id):
    
    # get the top 10 similar users
    _, indices = knn_model.kneighbors(combined_score_matrix_cosine_and_leniency_df.loc[[user_id]])
    similar_users = combined_score_matrix_cosine_and_leniency_df.iloc[indices[0], 1:]
    
    
    # get the not watched movies of the user
    not_watched = ratings_matrix.loc[user_id][ratings_matrix.loc[user_id] == 0].index
    
    # create a dictionary to store the score of each not watched movie
    movie_scores = {}
    
     # iterate over each similar user
    for user in similar_users.index:
        # get the ratings of the user for not watched movies
        user_ratings = ratings_matrix.loc[user][not_watched]

        # calculate the cosine similarity score between the user and input user
        similarity_score = combined_score_matrix_cosine_and_leniency_df[user_id][user]

        # multiply the ratings with the similarity score and add to the movie_scores dictionary
        for movie, rating in user_ratings.items():
            if movie in movie_scores:
                movie_scores[movie] += rating * similarity_score
            else:
                movie_scores[movie] = rating * similarity_score
    
    # sort the movie_scores dictionary by value in descending order
    sorted_scores = sorted(movie_scores.items(), key=lambda x: x[1], reverse=True)
    
    # get the top 5 movies from the sorted_scores
    top_5_movies = [x[0] for x in sorted_scores[:5]]
    
    return top_5_movies
    

userId = 1
top_5_recommended_movies = recommend_movies(userId)
movie_titles = complete_df.loc[complete_df['movieId'].isin(top_5_recommended_movies), 'original_title'].tolist()
print(f"Recommended movies for user {userId} are: ")
print()
unique_movie_titles = set(movie_titles)
for titles in unique_movie_titles:
    print(titles)
```
